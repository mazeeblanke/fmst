# 3) Feature Engineering techniques

We are going to extract a new feature age_months

```python
#Binning days to months
def label(data):
    if data["age"] <= 30:
        return "1 month"
    if data["age"] > 30 and data["age"] <= 60 :
        return "2 months"
    if data["age"] > 60 and data["age"] <= 90 :
        return "3 months"
    if data["age"] > 90 and data["age"] <= 120 :
        return "4 months"
    if data["age"] > 120 and data["age"] <= 150 :
        return "5 months"
    if data["age"] > 150 and data["age"] <= 180 :
        return "6 months"
    if data["age"] > 180 and data["age"] <= 210 :
        return "7 months"
    if data["age"] > 210 and data["age"] <= 240 :
        return "8 months"
    if data["age"] > 240 and data["age"] <= 270 :
        return "9 months"
    if data["age"] > 270 and data["age"] <= 300 :
        return "10 months"
    if data["age"] > 300 and data["age"] <= 330 :
        return "11 months"
    if data["age"] > 330 :
        return "12 months"
df["age_months"] = df.apply(lambda data:label(data) , axis=1)
```

So we are going to check the correlation between features, and if any has strong correlation we will drop one pf it

```python
print df.corr()
```

From our inspection, water and superplastic have some sort of correlation, so will drop superplastic

### 1) Creating the model and tuning it

Algorithms suitable for this project include:
    
* Linear Regression
* Gradient boost
* Ada Boost
* Random Forest

```python
target = "strength"
model_comp = pd.DataFrame(index=["ROOT MEAN SQUARED ERROR", "cv-mean", "cv-std", "cv-max", "cv-min", "Score without CV (train)", "SCORE without CV (test)"])
def model(algorithm,dtrainx,dtrainy,dtestx,dtesty,of_type, type):
    
    print "Algorithm in use: ", algorithm, '\n\n'
    print "***************************************************************************"
    algorithm.fit(dtrainx,dtrainy)
    prediction = algorithm.predict(dtestx)
    print "ROOT MEAN SQUARED ERROR :", np.sqrt(mean_squared_error(dtesty,prediction))
    print "***************************************************************************"
    prediction = pd.DataFrame(prediction)
    cross_val = cross_val_score(algorithm,dtrainx,dtrainy,cv=20)
    cross_val = cross_val.ravel()
    print "SCORE without CV (train):", algorithm.score(dtrainx,dtrainy) 
    print "SCORE without CV (test):", algorithm.score(dtestx,dtesty)
    print "CROSS VALIDATION SCORE"
    print "************************"
    print "cv-mean :",cross_val.mean()
    print "cv-std  :",cross_val.std()
    print "cv-max  :",cross_val.max()
    print "cv-min  :",cross_val.min()
    
    model_comp[type] = [
        np.sqrt(mean_squared_error(dtesty,prediction)),
        cross_val.mean(),
        cross_val.std(),
        cross_val.max(),
        cross_val.min(),
        algorithm.score(dtrainx,dtrainy),
        algorithm.score(dtestx,dtesty)
    ]
    
    plt.figure(figsize=(13,28))
    plt.subplot(211)
    
    testy = dtesty.reset_index()["strength"]
    
    ax = testy.plot(label="originals",figsize=(12,13),linewidth=2)
    ax = prediction[0].plot(label = "predictions",figsize=(12,13),linewidth=2)

    plt.legend(loc="best")
    plt.title("ORIGINALS VS PREDICTIONS")
    plt.xlabel("index")
    plt.ylabel("values")
    ax.set_facecolor("k")
    
    plt.subplot(212)
    
    if of_type == "coef":
        coef = pd.DataFrame(algorithm.coef_.ravel())
        coef["feat"] = dtrainx.columns
        ax1 = sns.barplot(coef["feat"],coef[0],palette="jet_r",
                          linewidth=2,edgecolor="k"*coef["feat"].nunique())
        ax1.set_facecolor("lightgrey")
        ax1.axhline(0,color="k",linewidth=2)
        plt.ylabel("coefficients")
        plt.xlabel("features")
        plt.title('FEATURE IMPORTANCES')
    
    elif of_type == "feat":
        coef = pd.DataFrame(algorithm.feature_importances_)
        coef["feat"] = dtrainx.columns
        ax2 = sns.barplot(coef["feat"],coef[0],palette="jet_r",
                          linewidth=2,edgecolor="k"*coef["feat"].nunique())
        ax2.set_facecolor("lightgrey")
        ax2.axhline(0,color="k",linewidth=2)
        plt.ylabel("coefficients")
        plt.xlabel("features")
        plt.title('FEATURE IMPORTANCES')

import warnings
warnings.filterwarnings("ignore")
```

```python
lr = LinearRegression()
model(lr,X_train,y_train,X_test,y_test,"coef", 'linear regression')
```

```python
from sklearn.ensemble import AdaBoostRegressor
adb = AdaBoostRegressor()
model(adb,X_train,y_train,X_test,y_test,"feat", "Ada boost")
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor()
model(gbr,X_train,y_train,X_test,y_test,"feat", "Gradient boost")
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model(rf,X_train,y_train,X_test,y_test,"feat", "Random Forest Regressor")
```

2) Techniques employed to squeeze that extra performance out of the model without making it overfit

```python
hyper_parameter_comp = pd.DataFrame(index=['cv-mean', 'cv-std', 'cv-max', 'cv-min'])
```

### Parameter tuning using gridsearch on Gradient boost estimator

```python
parameters = {
 'max_depth': [10, 20, 30, 40, 50],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4, 8],
 'n_estimators': [100]}

clf = GridSearchCV(gbr, parameters, cv = 5, verbose = 2, n_jobs= 4)

clf.fit(X_train, y_train)

clf.best_params_
```

```python
clf = GradientBoostingRegressor(max_depth= 20,
 max_features= 'sqrt',
 min_samples_leaf= 8,
 n_estimators= 100)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
cross_val_score(clf, X_train, y_train, cv = 5).mean()

val = cross_val_score(clf, X, y, cv = 5)
hyper_parameter_comp['Gradient boost'] = [
    val.mean(),
    val.std(),
    val.max(),
    val.min()
]
```

using parameter tuning we were able to increase the validation score to 93.22

### Parameter tuning using gridsearch on Random Forest Regressor

```python
parameters = {'bootstrap': [True],
 'max_depth': [10, 20, 30, 40, 50],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4, 8],
 'n_estimators': [100]}


clf = GridSearchCV(RandomForestRegressor(), parameters, cv = 5, verbose = 2, n_jobs= 4)
clf.fit(X, y)

clf.best_params_
```

```python
clf = RandomForestRegressor(bootstrap= True,
 max_depth=40,                           
 max_features= 'auto',
 min_samples_leaf= 1,
 n_estimators= 100)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
cross_val_score(clf, X_train, y_train, cv = 5).mean()

val = cross_val_score(clf, X_train, y_train, cv = 5)
hyper_parameter_comp['Random forest regressor'] = [
    val.mean(),
    val.std(),
    val.max(),
    val.min()
]
```
