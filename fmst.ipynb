```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('concrete.csv')
```

```python
independent_variables = df.drop(columns=['strength'])
dependent_variable = df[['strength']]
```

# Exploratory data quality report

## 1. Univariate analysis

### Inspect data type

```python
independent_variables.dtypes
```

Based on the data types listed we can see that all features/columns are of type float except age and that's reasonable because age should be an integer

### Check the info

```python
independent_variables.info()
```

The information from the info tells us that non of the features has a null value, although we will still check further to be sure

Next we will check the description of the independent attributes which should include (name, range of values observed, central values (mean and median), standard deviation and quartiles

```python
independent_variables.describe().transpose()
```

From the information above, we can see that there are 1030 rows in the dataset. the minimum age is 1 and maximum is 365, which is correct according to the data description.

### Checking for missing values

```python
independent_variables.isnull().any()
```

This information verifies what we got from the .info() method. there is no missing value in the dataset

### Checking for duplicates

```python
independent_variables.nunique()
```

Since all the columns do not have 1030 unique values, there are some duplicates

```python
for feature in list(independent_variables.columns):
    print "The value counts for " + feature + " is:" 
    print independent_variables[feature].value_counts()
    print "\n\n"
```

### Next we will detect outliers

```python
for feature in list(independent_variables.columns):
    print "Box plot for " + feature
    sns.boxplot(x=independent_variables[feature])
    plt.show()
    print "\n\n"
```

Based off the information above, we can see that:
    
    age has outliers between values 150 and 370
    
    fineagg has one outlier
    
    coarseagg has no outlier
    
    superplastic has two outliers between 25 and 70
    
    water has an outlier to the left hand side and two outliers to the right hand side
    
    ash has no outlier
    
    slag has one outlier
    
    cement has no outlier

### Analysis of the body of distributions

```python
for feature in list(independent_variables.columns):
    print "Box plot for " + feature
    sns.distplot(independent_variables[feature], kde=True)
    plt.show()
    print "\n\n"
```
